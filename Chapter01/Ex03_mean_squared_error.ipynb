#### MSE - mean squared error(평균 제곱 오차) 
$$ MSE = \frac{1}{n}\sum{(\hat{y_i}-y_i})^2$$

```python
fake_a_b = [3, 76]
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
def predict(x):
    return fake_a_b[0]*x + fake_a_b[1]

def mse(y_hat, y):
    return ((y_hat -y) ** 2).mean()

def mse_val(predict_reuslt, y):
    return mse(np.array(predict_reuslt), np.array(y))
```

```python
predict_result = []

for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print("Study= %f, real score = %.f, predict score = %.f" %(x[i], y[i], predict(x[i])))

print("mse value : " + str(mse_val(predict_result, y)))
```
