```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

#### method of least squares(최고제곱법)
$$ a = \frac{(x-x평균)(y-y평균)\text{의 합}}{(x-x평균)^2\text{의 합}}$$ 

```python
x = [2, 4, 6, 8]
y = [81, 93, 91, 97]
mean_x = np.mean(x)
mean_y = np.mean(y)
print(f"mean of x {mean_x}")
print(f"mean of y {mean_y}")
```

```python
def get_show_scatter(x, y):
    plt.figure(figsize=(12, 8))
    plt.scatter(x, y, color="blue")
    plt.show()
```

```python
get_show_scatter(x, y)
```

```python
# divisor = sum([(i-mean_x) **2 for i in x])
def get_divisor(x, maen_x):
    divisor = 0 
    
    for i in range(len(x)):
        result= ((x[i]-mean_x) **2) 
        divisor += result
    
    return divisor  

def get_dividend(x, mx, y, my):
    dividend = 0 
    for i in range(len(x)):
        result = (x[i]-mx) * (y[i]-my)
        dividend += result
    return dividend 
```

```python
divisor = get_divisor(x, mean_x)
dividend = get_dividend(x, mean_x, y, mean_y)
print(divisor)
print(dividend)
```

```python
 a = (dividend/divisor)
 b = (mean_y - (mean_x * a))
 print(f"a : {a}")
 print(f"b : {b}")
```

```python
reg_y = [] 
for i in x : 
    reg_y.append(a * i + b)
    print(a*i+b)
```
