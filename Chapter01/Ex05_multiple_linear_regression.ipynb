```python
data = [[2, 1, 81], [4, 4, 93], [6, 2, 91], [8 ,3, 97]]
x_1 = [i[0] for i in data]
x_2 = [i[1] for i in data]
y = [i[2] for i in data]
print(x_1)
print(x_2)
print(y)
```

```python
def get_show_scatter3d(x1, x2, y):
    ax = plt.axes(projection="3d")
    ax.set_xlabel("study_hour")
    ax.set_ylabel("private_classs")
    ax.set_zlabel("score")
    ax.scatter(x1, x2, y)
    plt.show()
```

```python
np_x1 = np.array(x_1)
np_x2 = np.array(x_2)
np_y = np.array(y)
```

```python

descent_x1 = 0 
descent_x2 = 0 
y_intercept = 0 
learning_rate = 0.005
epochs=2022
    
```

```python
def get_multiple_linear_dsecent_gradient(a1, a2, b, epochs, lr):
    for i in range(epochs):
        pred_y = (a1 * np_x1) + (a2 * np_x2) + b 
        loss  = np_y - pred_y 

        a1_diff = -(1/len(np_x1)) * sum(np_x1 * (loss))
        a2_diff = -(1/len(np_x2)) * sum(np_x2 * (loss))

        update_b = -(1/len(np_x1)) * sum(np_y - pred_y)

        a1 = a1- lr * a1_diff 
        a2 = a2- lr * a2_diff 

        b = b - (lr * update_b)

        if i % 100 == 0 :
            print(f"Epoch {i} {'===='*20}")
            print(f"gradient1 : {a1} \ngradient2 : {a2} \nY-Intercept : {b}" )
```

```python
descent_x1 = 0 
descent_x2 = 0 
y_intercept = 0 
learning_rate = 0.005
epochs=2022

get_multiple_linear_dsecent_gradient(descent_x1, descent_x2, y_intercept, epochs=2022, lr=0.005)

    
```
